```python
import pandas as pd
```

```python
path = r"D:\Python Files\Jupyter Notebook\Orielly\Housing Linear Regression\datasets\housing\housing.csv"
```

```python
housing = pd.read_csv(path)
```

```python
housing.head()
```

```python
import sklearn
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(housing, test_size = 0.2, random_state = 42)
```

# Data Prep

```python
housing = train_set.drop("median_house_value", axis = 1)
```

```python
housing_labels = train_set["median_house_value"].copy()
```

```python
housing_num = housing.drop("ocean_proximity", axis = 1)
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline = Pipeline([
                        ("imputer", SimpleImputer(strategy = "median")),
                        ("std_scaler", StandardScaler()),
                        ])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
num_attribs = list(housing_num)
```

```python
cat_attribs = ["ocean_proximity"]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])
```

```python
housing_prepared = full_pipeline.fit_transform(housing)
```

```python
import pickle
with open('housing_transformer.pkl', 'wb') as file:
    pickle.dump(full_pipeline, file)
```

# Model

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
```

```python
forest_reg.fit(housing_prepared, housing_labels)
```

```python
with open('model.pkl', 'wb') as file:
    pickle.dump(forest_reg, file)
```
